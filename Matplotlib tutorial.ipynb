```python
# handling multiple plots
# subplot(211) -> means that we have 2 plots vertically and 1 plot horizontally and this plot should be the first plot
# subplot(212) -> means that we have 2 plots vertically and 1 plot horizontally and this plot should be the second plot
import numpy as np
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2))

plt.subplot(212)
plt.plot(t2, np.cos(2 * np.pi * t2))
plt.show();
```
